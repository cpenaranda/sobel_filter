```python
import numpy as np
from PIL import Image, ImageOps
```

```python
sobel_x_operator = np.array([[1, 0, -1],
                             [2, 0, -2],
                             [1, 0, -1]])
sobel_y_operator = np.array([[ 1,  2,  1],
                             [ 0,  0,  0],
                             [-1, -2, -1]])
```

```python
def sobel_filter(name_image, x, y):
  image = np.asarray(ImageOps.grayscale(Image.open(name_image)), dtype=np.int32)
  image = np.pad(image, pad_width = 1, constant_values = 0)
  G_x = (x[0,0] * image[:-2, :-2]) + (x[0,1] * image[:-2, 1:-1]) + (x[0,2] * image[:-2, 2:]) + (x[1,0] * image[1:-1, :-2]) + (x[1,1] * image[1:-1, 1:-1]) + (x[1,2] * image[1:-1, 2:]) + (x[2,0] * image[2:, :-2]) + (x[2,1] * image[2:, 1:-1]) + (x[2,2] * image[2:, 2:])
  G_y = (y[0,0] * image[:-2, :-2]) + (y[0,1] * image[:-2, 1:-1]) + (y[0,2] * image[:-2, 2:]) + (y[1,0] * image[1:-1, :-2]) + (y[1,1] * image[1:-1, 1:-1]) + (y[1,2] * image[1:-1, 2:]) + (y[2,0] * image[2:, :-2]) + (y[2,1] * image[2:, 1:-1]) + (y[2,2] * image[2:, 2:])
  total = np.sqrt(G_x**2 + G_y**2)
  return total, G_x, G_y
```

```python
total, G_x, G_y = sobel_filter('../images/bike.jpg', sobel_x_operator, sobel_y_operator)
```

```python
Image.open('../images/bike.jpg').show()
for item in [G_x, G_y, total]:
  Image.fromarray(item).show()
```
